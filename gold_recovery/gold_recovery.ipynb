**Тех.процесс:**
    
1. Флотация
    
    Во флотационную установку подаётся смесь золотосодержащей руды. После обогащения получается черновой концентрат и «отвальные хвосты», то есть остатки продукта с низкой концентрацией ценных металлов.
    
    На стабильность этого процесса влияет непостоянное и неоптимальное физико-химическое состояние флотационной пульпы (смеси твёрдых частиц и жидкости).
    

2. Очистка
    
    Черновой концентрат проходит две очистки. На выходе получается финальный концентрат и новые отвальные хвосты.

**Описание данных**

- *Rougher feed* — исходное сырье
- *Rougher additions* (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
- *Xanthate* — ксантогенат (промотер, или активатор флотации)
- *Sulphate* — сульфат (на данном производстве сульфид натрия)
- *Depressant* — депрессант (силикат натрия)
- *Rougher process* — флотация
- *Rougher tails* — отвальные хвосты
- *Float banks* — флотационная установка
- *Cleaner process* — очистка
- *Rougher Au* — черновой концентрат золота
- *Final Au* — финальный концентрат золота

**Параметры этапов**

- *air amount* — объём воздуха
- *fluid levels* — уровень жидкости
- *feed size* — размер гранул сырья
- *feed rate* — скорость подачи

**Наименование признаков**

Наименование признаков должно быть такое: `[этап].[тип_параметра].[название_параметра]`

Возможные значения для блока `[этап]`:

- *rougher* — флотация
- *primary_cleaner* — первичная очистка
- *secondary_cleaner* — вторичная очистка
- *final* — финальные характеристики

Возможные значения для блока `[тип_параметра]`:

- *input* — параметры сырья
- *output* — параметры продукта
- *state* — параметры, характеризующие текущее состояние этапа
- *calculation* — расчётные характеристики

**Нужно спрогнозировать две величины:**

- эффективность обогащения чернового концентрата `rougher.output.recovery`
- эффективность обогащения финального концентрата `final.output.recovery`

$Recovery = {C * (F - T) \over F * (C - T)} * 100\%$

- *C* — доля золота в концентрате после флотации/очистки
- *F* — доля золота в сырье/концентрате до флотации/очистки
- *T* — доля золота в отвальных хвостах после флотации/очистки

Для прогноза коэффициента нужно найти долю золота в концентратах и хвостах. 

**Итоговая метрика** - симметричное среднее абсолютное процентное отклонение:

$sMAPE = {1 \over N} \sum_{i = 1}^{N}{|y_{i} - ŷ_{i}| \over (|y_{i}| + |ŷ_{i}|) / 2 } * 100\%$

$sMAPE_{total} = 25\% * sMAPE_{rougher} + 75\% * sMAPE_{final}$

### 1. 2 Проверка значений

Проверим, правильно ли посчитаны значения *rougher.output.recovery* в обучающей выборке, воспользовавшись формулой:

$Recovery = {C * (F - T) \over F * (C - T)} * 100\%$

- *C* — доля золота в концентрате после флотации - *rougher.output.concentrate_au*
- *F* — доля золота в сырье до флотации - *rougher.input.feed_au*
- *T* — доля золота в отвальных хвостах после флотации - *rougher.output.tail_au*

В этих столбцах нет пропущенных значений, поэтому можно сразу переходить к расчётам.

```python
def recovery (f, c, t):
    return c * (f - t) / (f * (c - t)) * 100

feed_train = gold_recovery['train']['rougher.input.feed_au']
c_rougher_train = gold_recovery['train']['rougher.output.concentrate_au']
t_rougher_train = gold_recovery['train']['rougher.output.tail_au']

print('Для расчётной и фактической эффективности обогащения после флотации MAE = {:.1e}'
      .format(mean_absolute_error(recovery(feed_train, c_rougher_train, t_rougher_train),
                                  gold_recovery['train']['rougher.output.recovery']
                                 )
             )
     )
```

Аналогично проверим final.output.recovery.

```python
c_final_train = gold_recovery['train']['final.output.concentrate_au']
t_final_train = gold_recovery['train']['final.output.tail_au']

print('Для расчётной и фактической эффективности обогащения после очистки MAE = {:.1e}'
      .format(mean_absolute_error(recovery(feed_train, c_final_train, t_final_train),
                                  gold_recovery['train']['final.output.recovery']
                                 )
             )
     )
```

Среднее абсолютное отклонение данных эффиктивности в таблице от расчётных значений фактически равно нулю. Данные об объёмах золота в таблице не противоречат сами себе.

Среди значений объёмов воздуха есть отрицательные числа, они близки к нулю. Возможно, это ошибка показаний датчика, связанная с его погрешностью. Тогда отрицательные значения нужно считать как 0. При этом погрешность есть для любого значения: также как показания 0 оказываются несколько выше или ниже нуля, любые показания n не будут абсолютно равны n. Поэтому если заменить отрицательные значения на 0, показание 0 окажется более точным, чем любое другое n. Для дальнейшего ислледования можно оставить данные в таком виде, как они были считаны с датчиков, и принимать решение об этих значениях уже после анализа зависимостей.

Уровень жидкости весь представлен отрицательными значениями. Либо это ошибка выгрузки данных, и "минус" означает "тире", либо уровень отсчитывается от максимального значения вниз, и тогда -800 - действительно наименьший уровень. Т.к. при выгрузке других данных не наблюдается тотальная замена положительных значений отрицательными, будем считать, что здесь тоже нет ошибки.

Среди расчётных значений тоже есть отрицательные. Судя по названиям столбцов, для расчётов использованы данные о золоте, сульфате и свинце, которые все положительные. Но не зная точных формул для расчётов, нельзя сказать, есть ли действительно ошибки в этих данных. Проверим отношение feed_au и feed_pb там, где это возможно, и сравним с au_pb_ratio.

```python
def pruning (df, column, min_value, max_value):
    df = df.copy()
    df[column] = (
        df[column]
        .where(df[column] <= max_value, max_value)
        .where(df[column] >= min_value, min_value)
    )
    return df
```

## 3 Модель

### 3.1 Подготовка к обучению модели

Создадим функцию, для расчёта симметричного среднего абсолютного процентного отклонения:

$sMAPE = {1 \over N} \sum_{i = 1}^{N}{|y_{i} - ŷ_{i}| \over (|y_{i}| + |ŷ_{i}|) / 2 } * 100\%$

На основе функции создадим метрику для выбора лучшей модели. Чем меньше значение sMAPE, тем лучше. При таком условии make_scorer изменит знак посчитанного sMAPE, поэтому вычисленные на кросс-валидации значения метрики будут отрицательными.

```python
def smape(target, predicted):
    return ((abs(target - predicted)) / ((abs(target) + abs(predicted)) / 2)).sum() / len(target) * 100

smap_error = make_scorer(smape, greater_is_better=False)
```

Необходимо предсказать эффективность обогащения на двух стадиях производства: после флотации и после финальной очистки.

$sMAPE_{total} = 25\% * sMAPE_{rougher} + 75\% * sMAPE_{final}$

Поэтому целевыми признаками являются:

- *rougher.output.recovery*
- *final.output.recovery*

Для предсказания итоговых показателей флотации нужны данные только по флотации, потому что на эту стадию процесса не может влиять будущая очистка. Создадим для этого отдельную модель.

1. На основе данных для флотации предскажем rougher.output.recovery.
2. Используем все признаки из тестовой выборки для предсказания final.output.recovery.

Создадим выборки для обучения и тестирования моделей.

- *X_rougher_train* - признаки для обучения модели на стадии флотации: признаки, входящие в тестовую выборку и содержащие слово rougher.
- *y_rougher_train* - целевой признак rougher.output.recovery.


- *X_rougher_test* - признаки со словом rougher.
- *y_rougher_test* - значения rougher.output.recovery из таблицы full.


- *X_final_train* - все признаки, которые есть в таблице test.
- *y_final_train* - целевой признак final.output.recovery.


- *X_final_test* - все признаки тестовой выборки.
- *y_final_test* - значения final.output.recovery из таблицы full.

```python
features_rougher_columns = list(gold_recovery['test'].filter(like = 'rougher'))
features_final_columns = list(gold_recovery['test'])

gold_recovery['test'] = (
    gold_recovery['test']
    .join(gold_recovery['full'][['rougher.output.recovery', 'final.output.recovery']])
)
```

```python
X_rougher_train = gold_recovery['train'][features_rougher_columns]
y_rougher_train = gold_recovery['train']['rougher.output.recovery']

X_rougher_test = gold_recovery['test'][features_rougher_columns]
y_rougher_test = gold_recovery['test']['rougher.output.recovery']

X_final_train = gold_recovery['train'][features_final_columns]
y_final_train = gold_recovery['train']['final.output.recovery']

X_final_test = gold_recovery['test'][features_final_columns]
y_final_test = gold_recovery['test']['final.output.recovery']
```

Все необходимые выборки созданы.

Напишем функцию для получения предсказаний.

```python
def pred_recovery(model, X):
    pred = model.predict(X)
    pred = pd.Series(pred, index = X.index, name = 'recovery_predicted')
    
    return pred
```

**Вывод**

Для оценки предсказания эффективности обогащения создана функция, считающая симметричное среднее абсолютное процентное отклонение.

Для предсказаний эффективности обогащения золота после флотации и после финальной очистки решено использовать две разные модели. Для каждой модели созданы обучающие и тестовые выборки.

Подготовлена функция для будущих моделей, пердсказывающая эффективность.

### 3.2 Создание модели

Посмотрим, какое значение sMAPE даёт линейная регрессия на кросс-валидации.

```python
kf = KFold(n_splits=5, random_state=42, shuffle=True)

def lr_model(X, y):
    model = LinearRegression()
    smape_scores = cross_val_score(model, X, y, scoring = smap_error, cv = kf)
    smape_mean = abs(smape_scores.mean())
    
    print('sMAPE на кросс-валидации: {}'.format(np.round(abs(smape_scores), 2)))
    print('sMAPE_mean: {:.2f}%'.format(smape_mean))
    
    return model

model_lr_rougher = lr_model(X_rougher_train, y_rougher_train)
```

Рассмотрим модель случайного леса. Количество признаков достаточно большое, и поиск оптимальных гиперпараметров займёт время. Выделим самые важные для дерева решений признаки и используем их для случайного леса.

```python
model_dt_rougher = DecisionTreeRegressor()
model_dt_rougher.fit(X_rougher_train, y_rougher_train)
features_rougher = pd.DataFrame({'name': X_rougher_train.columns, 'importance': model_dt_rougher.feature_importances_})
features_rougher.sort_values(by='importance', ascending=False)
```

```python
alpha = 0.015

best_features_rougher = features_rougher[features_rougher['importance'] > alpha]['name'].values
```

Подберём лучшие гиперпараметры для леса. (В работе оставлена сетка на 4 варианта леса - изначально сетка была больше, но расчёты занимают время.)

```python
def rf_model(X, y, params):
    model = RandomForestRegressor(random_state = 42)
    grid_search = GridSearchCV(model, params, scoring = smap_error, cv = kf)
    grid_search.fit(X, y)
    smape_best = abs(grid_search.best_score_)

    print('Лучшие параметры:', grid_search.best_params_)
    print('sMAPE = {:.2f}%'.format(smape_best))
    
    return grid_search.best_estimator_

model_rf_rougher = rf_model(
    X_rougher_train[best_features_rougher], 
    y_rougher_train, 
    {
        'n_estimators': range(145, 155, 5), 
        'max_depth': range(25, 30, 3)
    }
)
```

Случайный лес даёт существенно лучшие показатели на обучающей выборке, чем линейная регрессия. Сохраним предсказания этой модели и значение sMAPE. 

```python
pred_rougher_train = pred_recovery(model_rf_rougher, X_rougher_train[best_features_rougher])
smape_r_train = smape(y_rougher_train, pred_rougher_train)

print('sMAPE эффективности после флотации: {:.2f}%'.format(smape_r_train))
```

Значения эффективности после флотации могут быть важным параметром в предсказывании финальной эффективности. Добавим столбец с предсказаниями первой модели к обучающей выборке.

```python
X_final_train = X_final_train.join(pred_rougher_train)
```

Теперь обучим модель линейной регрессии для прогнозирования эффективности после финальной очистки.

```python
model_lr_final = lr_model(X_final_train, y_final_train)
```

Как и на предыдущем шаге, выберем наиболее важные признаки для дерева решений.

```python
model_dt_final = DecisionTreeRegressor()
model_dt_final.fit(X_final_train, y_final_train)
features_final = pd.DataFrame({'name': X_final_train.columns, 'importance': model_dt_final.feature_importances_})
best_features_final = features_final[features_final['importance'] > alpha]['name'].values
```

Подберём гиперпараметры для случайного леса.

```python
model_rf_final = rf_model(
    X_final_train[best_features_final], 
    y_final_train, 
    {
        'n_estimators': range(155, 165, 5), 
        'max_depth': range(25, 30, 3)
    }
)
```

На полной выборке случайный лес даёт также лучшее значение метрики, чем линейная регрессия. Сохраним предсказания этой модели и значение sMAPE.

```python
pred_final_train = pred_recovery(model_rf_final, X_final_train[best_features_final])
smape_f_train = smape(y_final_train, pred_final_train)

print('sMAPE эффективности после финальной очистки: {:.2f}%'.format(smape_f_train))
```

Посчитаем $sMAPE_{total}$ на обучающей выборке.

```python
def smape_total(smape_r, smape_f):
    return 0.25 * smape_r + 0.75 * smape_f

smape_t_train = smape_total(smape_r_train, smape_f_train)

print('Итоговое значение sMAPE = {:.2f}%'.format(smape_t_train))
```

Создадим функцию для использования обеих моделей в связке. Не смотря на то, что у нас уже созданы тестовые выборки X_rougher_test и X_final_test, предусмотрим выделение подобных выборок внутри функции, тем более что количество признаков для случайного леса изменилось.

```python
def get_predictions(features):
    X_rougher = features[best_features_rougher]
    pred_rougher = pred_recovery(model_rf_rougher, X_rougher)
    
    X_final = features.join(pred_rougher)
    X_final = X_final[best_features_final]
    pred_final = pred_recovery(model_rf_final, X_final)
    
    return pred_rougher, pred_final
```

**Вывод**

Рассмотрены модели линейной регрессии и случайный лес с четырьмя наборами гиперпараметров. Лес обучен на сокращённом наборе наиболее важных для дерева признаков. Регрессия работает быстрее, но лес даёт лучшее значение sMAPE.

Отдельно для флотации и для всего тех.процесса в целом созданы соответствующие модели для предсказания эффективности. Итоговое значение sMAPE на обучающей выборке составило 2,4%.

Для получения предсказанных значений эффективности создана функция get_predictions.

### 3.3 Тестирование модели

```python
pred_rougher_test, pred_final_test = get_predictions(gold_recovery['test'])
```

Посчитаем метрики.

```python
smape_r_test = smape(y_rougher_test, pred_rougher_test)
smape_f_test = smape(y_final_test, pred_final_test)
smape_t_test = smape_total(smape_r_test, smape_f_test)

smape_test = pd.DataFrame(
    {
        'train':[smape_r_train, smape_f_train, smape_t_train],
        'test':[smape_r_test, smape_f_test, smape_t_test],
    }, 
    index = ['sMAPE_rougher', 'sMAPE_final', 'sMAPE_total']
)

smape_test.applymap(lambda x: '{:.2f}%'.format(x))
```

Значения метрики на тестовой выборке выше, чем на обучающей. Результаты работы леса на тесте сопоставимы с линейной регрессией на обучающей выборке. Модель, предсказывающая эффективность флотации, работает лучше, чем вторая модель.
